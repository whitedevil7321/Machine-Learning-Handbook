```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import joblib
```

```python
df=pd.read_csv("carprices.csv")
```

```python
df
```

```python
dummies=pd.get_dummies(df["Car_Model"],dtype="int")
```

```python
dummies
```

```python
dataFrame=pd.concat([dummies,df],axis="columns")
```

```python
dataFrame
```

```python
dataFrame.drop(["Mercedez_Benz_C_class","Car_Model"],inplace=True,axis=1)
```

```python
dataFrame
```

```python
x=dataFrame[["Audi A5","BMW X5","Mileage" ,"Age"]]
```

```python
x
```

```python
y=dataFrame[["Sell_Price"]]
```

```python
reg=linear_model.LinearRegression()
reg.fit(x,y)
```

```python
reg.predict([[0,0,70000,8]])
```

```python
joblib.dump(reg,"car_regression_model")
```

```python
model=joblib.load("car_regression_model")
```

```python
model.predict([[0,0,70000,8]])
```
