```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.preprocessing import LabelEncoder

```

```python
dataset=pd.read_csv("salaries.csv")
```

```python
dataset
```

```python
le=LabelEncoder()
dataset["company_n"]=le.fit_transform(dataset["company"])
```

```python
dataset
```

```python
dataset["job_n"]=le.fit_transform(dataset["job"])
```

```python
dataset
```

```python
dataset["degree_n"]=le.fit_transform(dataset["degree"])
```

```python
dataset=dataset.drop(["company","job","degree"],axis="columns")
```

```python
dataset
```

```python
y=dataset["salary_more_then_100k"]
```

```python
x=dataset[["company_n","job_n","degree_n"]]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.2)
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict([[2,0,0]])
```

```python
model.score(x_train,y_train)
```
