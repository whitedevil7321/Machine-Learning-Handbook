```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge,Lasso,LinearRegression
from sklearn.preprocessing import LabelEncoder
```

```python
dataset=pd.read_csv("Melbourne_housing_FULL.csv")
```

```python
dataset
```

```python
labelsofdata=["Suburb","Rooms","Type","Price","Method","Distance","Postcode","Bedroom2","Bathroom","Car","Landsize","BuildingArea","Propertycount","Regionname"]
```

```python
dataset=dataset[labelsofdata]
```

```python
dataset
```

```python
dataset.isna().sum()
```

```python
qq=["Bedroom2","Bathroom","Car","Landsize","BuildingArea"]
```

```python
dataset[qq]=dataset[qq].fillna(0)
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.isna().sum()
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
x=dataset.drop("Price",axis=1)
```

```python
y=dataset["Price"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
lasso=Lasso(alpha=50,max_iter=100,tol=0.1)
lasso.fit(x_train,y_train)
```

```python
lasso.score(x_test,y_test)
```

```python
lasso.score(x_train,y_train)
```

```python
Ridge=Ridge(alpha=50,max_iter=100,tol=0.1)
Ridge.fit(x_train,y_train)
```

```python
Ridge.score(x_train,y_train)
```

```python
Ridge.score(x_test,y_test)
```
