```python
import pandas as pd
```

```python
df=pd.read_csv("height_weight.csv")
```

```python
df.describe()
```

```python
Q1_height,Q3_height=df.height.quantile([0.25,0.75])
```

```python
Q1_height,Q3_height
```

```python
Q1_weight,Q3_weight=df.weight.quantile([0.25,0.75])
```

```python
Q1_weight,Q3_weight
```

```python
IQR_height=Q3_height-Q1_height
```

```python
IQR_weight=Q3_weight-Q1_weight
```

```python
lower_limit_height=Q1_height-(1.5*IQR_height)
```

```python
lower_limit_height
```

```python
upper_limit_height=Q3_height+(1.5*IQR_height)
```

```python
upper_limit_height
```

```python
lower_limit_weight=Q1_weight-(1.5*IQR_weight)
```

```python
lower_limit_weight
```

```python
upper_limit_weight=Q3_weight+(1.5*IQR_weight)
```

```python
upper_limit_weight
```

```python
df2=df[((df.height>lower_limit_height)&(df.height<upper_limit_height)) &((df.weight>lower_limit_weight)&(df.height<upper_limit_weight))]
```
