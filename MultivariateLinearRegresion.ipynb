```python
import pandas as pd
from word2number import w2n

# Assuming df['experience'] has a mixture of numeric and word values.
df['experience_str'] = df['experience'].astype(str)  # Convert to string

# Apply the word_to_num function to each row
def safe_word_to_num(val):
    try:
        return w2n.word_to_num(val)
    except ValueError:
        return None  # Or handle invalid entries differently, like returning NaN

df['experience_num'] = df['experience_str'].apply(safe_word_to_num)

# Now df['experience_num'] contains the numeric equivalents of worded numbers.
```

```python
df['experience']=df['experience_num']
df
```

```python
impute=SimpleImputer(missing_values=np.nan,strategy="most_frequent")
df['experience']=impute.fit_transform(df[['experience']])
```

```python
df.drop("experience_str",inplace=True,axis=1)
```

```python
mean_impute=SimpleImputer(missing_values=np.nan,strategy="mean")
```

```python
df['test_score']=mean_impute.fit_transform(df[['test_score']])
```

```python
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[["experience","test_score","score"]],df.salary)
```
